```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton, QVBoxLayout, QMessageBox, QDialog, QDesktopWidget, QHBoxLayout, QScrollArea
from PyQt5.QtGui import QFont
from PyQt5.QtCore import Qt

class AnaPencere(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle('Boru Fiyat Hesaplama')
        self.resize(400, 600)
        self.initUI()

    def initUI(self):
        buttons_layout = QVBoxLayout()
        buttons = []
        button_names = ["H8 Honlanmış Boru", "Krom Kaplı Mil", "H9 Honlanmış Boru", "İndiksiyonlu Mil", "Boru Rod"]
        for name in button_names:
            button = QPushButton(name)
            button.setFont(QFont('Arial', 12))
            button.setStyleSheet("""
                QPushButton {
                    background-color: #4CAF50; 
                    color: white; 
                    border-radius: 10px; 
                    padding: 10px;
                }
                QPushButton:hover {
                    background-color: #45a049;
                }
            """)
            button.setFixedHeight(50)
            buttons_layout.addWidget(button)
            buttons.append(button)

        layout = QVBoxLayout()
        layout.addLayout(buttons_layout)

        buttons[0].clicked.connect(lambda: self.open_cap_secim_penceresi("H8 Honlanmış Boru"))
        buttons[1].clicked.connect(lambda: self.open_cap_secim_penceresi("Krom Kaplı Mil"))
        buttons[2].clicked.connect(lambda: self.open_cap_secim_penceresi("H9 Honlanmış Boru"))
        buttons[3].clicked.connect(lambda: self.open_cap_secim_penceresi("İndiksiyonlu Mil"))
        buttons[4].clicked.connect(lambda: self.open_cap_secim_penceresi("Boru Rod"))

        self.setLayout(layout)
        self.center()

    def center(self):
        qr = self.frameGeometry()
        cp = QDesktopWidget().availableGeometry().center()
        qr.moveCenter(cp)
        self.move(qr.topLeft())

    def open_cap_secim_penceresi(self, boru_adi):
        if boru_adi in cap_fiyat:
            dialog = CapSecimPenceresi(boru_adi, cap_fiyat[boru_adi], self)
            dialog.exec_()
        else:
            QMessageBox.critical(self, 'Hata', f"{boru_adi} için çap bilgisi bulunamadı.")

class CapSecimPenceresi(QDialog):
    def __init__(self, boru_adi, cap_fiyati, parent=None):
        super().__init__(parent)
        self.setWindowTitle(f'{boru_adi} Çap Seçimi')
        self.setWindowFlags(Qt.FramelessWindowHint)
        self.setStyleSheet("background-color: #f0f0f0;")
        screen = QDesktopWidget().screenGeometry()
        self.setGeometry(0, 0, screen.width(), screen.height())
        
        scroll_area = QScrollArea()
        scroll_area.setWidgetResizable(True)
        scroll_widget = QWidget()
        scroll_layout = QVBoxLayout(scroll_widget)
        scroll_layout.setSpacing(20)

        self.boru_adi = boru_adi
        self.cap_fiyati = cap_fiyati

        for cap_aralik in self.cap_fiyati.keys():
            button = QPushButton(f'{cap_aralik[0]}-{cap_aralik[1]} mm')
            button.setFont(QFont('Arial', 11))
            button.setStyleSheet("""
                QPushButton {
                    background-color: #4CAF50; 
                    color: white; 
                    border-radius: 10px; 
                    padding: 10px;
                }
                QPushButton:hover {
                    background-color: #45a049;
                }
            """)
            button.setFixedSize(200, 50)
            button.clicked.connect(lambda _, cap_aralik=cap_aralik: self.show_price(cap_aralik))
            scroll_layout.addWidget(button)

        scroll_area.setWidget(scroll_widget)
        
        close_button = QPushButton("X")
        close_button.setStyleSheet("""
            QPushButton {
                background-color: #f44336; 
                color: white; 
                border-radius: 10px; 
                padding: 10px;
            }
            QPushButton:hover {
                background-color: #d32f2f;
            }
        """)
        close_button.setFixedSize(40, 40)
        close_button.clicked.connect(self.close)

        minimize_button = QPushButton("-")
        minimize_button.setStyleSheet("""
            QPushButton {
                background-color: #2196F3; 
                color: white; 
                border-radius:10px; 
                padding: 10px;
            }
            QPushButton:hover {
                background-color: #1976D2;
            }
        """)
        minimize_button.setFixedSize(40, 40)
        minimize_button.clicked.connect(self.showMinimized)

        header_layout = QHBoxLayout()
        header_layout.addStretch(1)
        header_layout.addWidget(minimize_button)
        header_layout.addWidget(close_button)
        
        layout = QVBoxLayout()
        layout.addLayout(header_layout)
        layout.addWidget(scroll_area)
        
        self.setLayout(layout)

    def show_price(self, cap_aralik):
        price = self.cap_fiyati[cap_aralik]
        QMessageBox.information(self, 'Fiyat', f"Çap {cap_aralik[0]}-{cap_aralik[1]} mm için {self.boru_adi} fiyatı: {price} Euro/mt")

    def center(self):
        qr = self.frameGeometry()
        cp = QDesktopWidget().availableGeometry().center()
        qr.moveCenter(cp)
        self.move(qr.topLeft())

# PyQt5 uygulaması oluşturma
app = QApplication(sys.argv)
window = AnaPencere()
window.showFullScreen()
sys.exit(app.exec_())
```
